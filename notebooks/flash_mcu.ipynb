```python
import utils
import keras
from keras.layers import Dense, Conv2D, Flatten
from src.ModelExecutor import convert_to_tflite, execute_model
import numpy as np
```

```python
model = keras.Sequential()
model.add(Conv2D(32, kernel_size=3, activation='relu', input_shape=(28,28,1)))
model.add(Flatten())
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
representative_data = np.random.rand(1, 28, 28, 1).astype(np.float32)

model_bytes = convert_to_tflite(
    model_or_path=model,
    output_path="harfm_model.tflite",
    representative_data=representative_data,
    npz_key=None,
    num_samples=2,
    sequence_length=None,
    quantization_mode="float16"
)

execute_model(model_bytes, port="/dev/ttyACM0", env="portenta_h7_m7")
```
